Описание полей:
*  StudentID – уникальный идентификатор для каждого студента.
*  Name – имя студента.

родителями студента (высокий/средний/низкий).
* FinalGrade: итоговая оценка студента (из 100), которая служит целевой переменной для прогнозирования.

# *7. Какие признаки в наибольшей степени предсказывают успеваемость студентов?*

```python
df_copy = df.copy()

# только с числовыми столбцами
df_copy = df_copy.select_dtypes(include=['number'])

#корреляция по финальной оценке
correlation = df_copy.corr()['FinalGrade'].sort_values(ascending=False)

for feature, corr in correlation.items():
    if feature != 'FinalGrade' and  feature != 'StudentID' :
        print(feature, ": ",corr)
df.head()
```

Наиболее предсказывают успеваемость студентов прошлый рейтинг, уровень посещаемости, количество часов посещения.

# *9. Оцените, оказывает ли участие во внеклассных мероприятиях положительное или отрицательное влияние на успеваемость*

```python
df_copy = df.copy()

with_activities = df_copy[df_copy['ExtracurricularActivities'] > 0]['FinalGrade'].mean()
without_activities = df_copy[df_copy['ExtracurricularActivities'] == 0]['FinalGrade'].mean()

print(" Средняя успеваемость с мероприятиями: ", with_activities)
print(" Средняя успеваемость без мероприятий: ", without_activities)

if with_activities>without_activities:
   print("Участие во внеклассных мероприятиях оказывает положительное влияние")
else:
  print("Участие во внеклассных мероприятиях оказывает отрицательное влияние")
```

# *10. Как различные уровни родительской поддержки влияют на успеваемость студентов?*

```python
result = df.groupby('ParentalSupport')['FinalGrade'].mean()
print("Средние оценки по уровням родительской поддержки:")
print(result)
```

Чем больше уровень поддержки, тем выше успеваемость.

# *11. Какой вклад в успеваемость вносит время, потраченное на учебу?*

```python
df_copy = df.copy()

low_study = df_copy[df_copy['StudyHoursPerWeek'] <= 10]['FinalGrade'].mean()
medium_study = df_copy[(df_copy['StudyHoursPerWeek'] > 10) & (df_copy['StudyHoursPerWeek'] <= 20)]['FinalGrade'].mean()
high_study = df_copy[df_copy['StudyHoursPerWeek'] > 20]['FinalGrade'].mean()

print("До 10 часов: ", low_study)
print("11-20 часов: ", medium_study)
print("Более 20 часов: ",high_study)
```

Чем больше количество часов, тем выше успеваемость.

# `12.Определите наиболее распространенное мужское и женское имя`

# *13.Проиндексируйте датафрейм по полю gender и определите раздельно наибольшее и наименьшее значение процента посещенных занятий*

```python
df_copy = df.copy()

max_attendance = df_copy.groupby('Gender')['AttendanceRate'].max()
min_attendance = df_copy.groupby('Gender')['AttendanceRate'].min()

print("Наибольшая посещаемость:")
print(max_attendance)
print("\nНаименьшая посещаемость:")
print(min_attendance)
```

# *14.Преобразуйте значения StudyHoursPerWeek в числа с плавающей запятой*

# *15.Вычислите для столбца StudyHoursPerWeek среднее значение, моду, медиану, дисперсию и среднеквадратичное отклонение*

```python
df_copy = df.copy()

study_hours = df_copy['StudyHoursPerWeek']

mean_value = study_hours.mean()
mode_value = study_hours.mode()[0]
median_value = study_hours.median()
variance_value = study_hours.var()
std_value = study_hours.std()

print("Статистические показатели StudyHoursPerWeek:")
print("Среднее значение: ", mean_value)
print("Мода: ", mode_value)
print("Медиана: ", median_value)
print("Дисперсия: ", variance_value)
print("Среднеквадратичное отклонение: ",std_value)
```
